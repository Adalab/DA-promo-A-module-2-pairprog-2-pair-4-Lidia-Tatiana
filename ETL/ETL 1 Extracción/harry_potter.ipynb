```python
import requests
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
#Creamos un diccionario vacio

url = "http://hp-api.herokuapp.com/api/characters/students"
response = requests.get(url=url)
response.status_code #200
response.text  #si me devuelve correcto
```

```python
#Nos guardamos el response en un data set
data = pd.json_normalize(response.json()) #dentro de json podemos poner "rows"
data
```
